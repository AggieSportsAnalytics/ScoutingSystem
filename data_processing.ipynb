```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("CollegeBasketballPlayers2009-2021.csv")
df.head()
```

```python
min_minutes_played = 5
```

```python
df = pd.DataFrame(df[df['Min_per'] > min_minutes_played])
df.head()
```

```python
positions = df['Unnamed: 64']
```

```python
positions.unique()
```

```python
df = df.rename({'Unnamed: 64': 'position'}, axis=1)
```

```python
df['position'].unique()
```

```python
positions = []
for x in df['position'].unique():
    positions.append(x)
positions = positions[1:]
positions
```

```python
df[df['position'] == 'Combo G'].to_csv('combo-g.csv')
df[df['position'] == 'Wing F'].to_csv('wing-f.csv')
```

```python
df[df['position'] == 'Stretch 4'].to_csv('split_by_position/stretch-4.csv')
```

```python
for i in range(0, len(positions)):
    pos = positions[i]
    pos = pos.lower()
    pos = '-'.join(pos.split())
    positions[i] = pos
```
